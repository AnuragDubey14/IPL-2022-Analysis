### Number of matches won by each team in ipl 2022

```python
x=df['match_winner'].value_counts()
x
```

```python
ax=plt.bar(x.index,x,color='#4CB391')
plt.xticks(rotation=60)
plt.title("Number of Matches by Team")

plt.show()
```

```python
won_by=df['won_by'].value_counts()
won_by
plt.pie(won_by,labels=won_by.index,colors=['blue','pink'])
plt.title("Number of matches won by chasing or defending")
plt.show()
```

```python
bowler=df['best_bowling'].value_counts()
plt.figure(figsize=(15,5))
plt.bar(bowler.index,bowler,color='#51e2f5')
plt.xticks(rotation=90)
plt.title("Best Bowler in IPL 2022")
plt.show()
```

```python
man_ofthematch=df['player_of_the_match'].value_counts()
plt.figure(figsize=(15,7))
plt.bar(man_ofthematch.index,man_ofthematch,color='lightgreen')
plt.title("Number of time player of the match")
plt.xticks(rotation=90)
plt.show()
```

### Top Scorer in IPL 2022

```python
plt.figure(figsize=(15,7))
plt.bar(df['top_scorer'],df['highscore'],color='m')
plt.title("Top Scorer in IPL 2022")
plt.xticks(rotation=90)
plt.show()
```
